# Smart Audio Enhancement & ANC (Offline Demo)
This notebook shows how to run the offline pipeline on a WAV file and plot waveforms/spectrograms.

```python
from smart_audio.pipeline_offline import offline_process
from smart_audio.utils.plotting import plot_waveform, plot_spectrogram
from smart_audio.utils.io_utils import load_audio
import matplotlib.pyplot as plt

primary_path = "audio_sample/sample_audio.wav"  # Replace with your file
reference_path = "audio_sample/noise_sample.wav"  # Optional: path to reference noise mic wav

y, anc_out, enh_out, sr = offline_process(primary_path, reference_path, sr=16000,
                                          anc_type='nlms', filter_len=256, mu=0.3,
                                          enhancement='wiener', out_prefix='demo')

plot_waveform(y, sr, 'Input waveform (primary)')
plot_spectrogram(y, sr, title='Input spectrogram (primary)')

plot_waveform(anc_out, sr, 'After ANC (error signal)')
plot_spectrogram(anc_out, sr, title='After ANC spectrogram')

plot_waveform(enh_out, sr, 'After Enhancement (final)')
plot_spectrogram(enh_out, sr, title='After Enhancement spectrogram')

plt.show()
```
